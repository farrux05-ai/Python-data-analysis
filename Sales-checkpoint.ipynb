```python
# KUN 1: SALES ANALYZER
# Maqsad: List comprehension, lambda, dictionary mastery

# Ma'lumotlar - bir haftalik sotuvlar
sales_data = [
    {'day': 'Dushanba', 'product': 'Non', 'quantity': 50, 'price': 2000},
    {'day': 'Dushanba', 'product': 'Sut', 'quantity': 30, 'price': 5000},
    {'day': 'Seshanba', 'product': 'Non', 'quantity': 45, 'price': 2000},
    {'day': 'Seshanba', 'product': 'Tuxum', 'quantity': 20, 'price': 1500},
    {'day': 'Chorshanba', 'product': 'Non', 'quantity': 60, 'price': 2000},
    {'day': 'Chorshanba', 'product': 'Sut', 'quantity': 25, 'price': 5000},
    {'day': 'Payshanba', 'product': 'Tuxum', 'quantity': 35, 'price': 1500},
    {'day': 'Juma', 'product': 'Non', 'quantity': 80, 'price': 2000},
    {'day': 'Juma', 'product': 'Sut', 'quantity': 40, 'price': 5000},
    {'day': 'Shanba', 'product': 'Non', 'quantity': 70, 'price': 2000},
]
```

```python
# ============================================
# VAZIFA 1: Har bir sotuvning umumiy summasini hisoblang
# O'rganasiz: List comprehension, dictionary access
# ============================================
def total_add(sales):
    """ Bu funksiya sales datasini olib uning ichidagi quantity va priceni ko'payitirib revenue deagn yangi ustunni qaytaradi"""
    return[{**sale, 'revenue': sale['quantity'] * sale['price']}
          for sale in sales] # bu  yerda ** sale yangi ro'yxat ochadi keyin unga yangi revenue degan ustun qo'shadi. 
enriched_list = total_add(sales_data)
print(f"Birinchi vazifa muvaffaqiyatli bajarildi. Natija: {enriched_list[0]}")
```

```python
# ============================================
# VAZIFA 2: Eng ko'p sotilgan mahsulotni toping
# O'rganasiz: filter, max, lambda
# ============================================
def best_seller_product(sales):
    """Eng ko'p MIQDORDA sotilgan mahsulotni topadi"""
    product_totals = {}
    for sale in sales:
        product = sale['product']
        product_totals[product] = product_totals.get(product, 0) + sale['quantity']# get bizga bu yerda qiymatlarni qaytaradi agar  yo'q bo'lsa xato bermaydi, 0 qo'shib ketaveradi
        
    best = max(product_totals, key=product_totals.get) # bu yerda agar sikldan ajratilmasa bitta aylanishda javob beradi. ya'ni birinchidagi nonni qaytaradi
    return best, product_totals[best]
best_product, total_qty = best_seller_product(sales_data)
print(f"\nâœ… VAZIFA 2: Eng ko'p sotilgan - {best_product}: {total_qty} dona")

# kam sotilgan mahsulot
def less_seller_product(sales):
    """ Bu eng kam sotilgan mahsulotni qaytaradi"""
    less_seller = {}
    for sale in sales:
        product = sale['product']
        less_seller[product] = less_seller.get(product, 0) + sale['quantity']
        
    less = min(less_seller, key=lambda x: less_seller[x])
    return less, less_seller[less]
print(f"Eng kam sotilgan mahsulot: ",less_seller_product(sales_data))

#eng ko'p daromad keltirgan mahsulot
def best_revenue_product(sales):
    """ Bu  funksiya eng ko'p qaysi mahsulot daromad keltirganini ko'rsatadi"""
    best_revenue = {}
    for sale in sales:
        product = sale['product']
        best_revenue[product] = best_revenue.get(product, 0) + sale['revenue']
    best = max(best_revenue, key = best_revenue.get)
    return best, best_revenue[best]
print(f"Eng ko'p sotillgan mahsulot: {best_revenue_product(enriched_list)}")
```

```python
# ============================================
# VAZIFA 3: Kunlik umumiy daromadni hisoblang
# O'rganasiz: Grouping, dictionary operations
def daily_revenue(sales):
    """ Bu funksiya kunlik daromadni ko'rsatadi"""
    daily = {}
    for sale in sales:
        day = sale['day']
        revenue = sale['price'] * sale['quantity']
        daily[day] = daily.get(day,0) + revenue
    return daily
    
daily_report = daily_revenue(sales_data)
print("\n Vazifa bajarildi. Kunlik daromad: ")
for day, revenue in daily_report.items():
    print(f"{day} kuni daromad : {revenue} so'm")
# Eng ko'p savdo bo'lgan kun
def best_selling_day(sales):
    """ BU funksiya eng yaxshi savdo bo'lgan kunni ko'rsatadi"""
    best_selling = {}
    for sale in sales:
        day = sale['day']
        revenue = sale['price'] * sale['quantity']
        best_selling[day] = best_selling.get(day,0) + revenue
        
    best_day = max(best_selling, key = best_selling.get)
    return best_day, best_selling[best_day]
high_revenue = best_selling_day(sales_data)
print(f"Eng ko'p daromad keltirgan kun {high_revenue[0]} : savdo {high_revenue[1]} so'm")
```

```python
# ============================================
# VAZIFA 4: Eng yaxshi va eng yomon kunni aniqlang
# O'rganasiz: max/min with custom key
# ============================================
def best_and_worst_day(daily_report):
    """ Bu eng yaxshi va eng yomon sotilgan kunni ko'rsatadi"""
    best = max(daily_report, key=daily_report.get)
    worst = min(daily_report, key=daily_report.get)

    return{
        'best' : (best, daily_report[best]),
        'worst':( worst, daily_report[worst])
    }
result = best_and_worst_day(daily_report)
print(f"\nVazifa 4 bajarildi")
print(f"Eng yuqori savdo bo'lgan kun {result['best'][0]} - {result['best'][1]} so'm")
print(f"Eng yomon savdo bo'lgan kun {result['worst'][0]} - {result['worst'][1]} so'm")
```

```python
# ============================================
# VAZIFA 5: Mahsulotlarning o'rtacha narxini hisoblang
# O'rganasiz: Aggregation, average calculation
# ============================================
def product_averages(sales):
    product_data = {}
    for sale in sales:
        product = sale['product']
        if product not in product_data:
            product_data[product] = {'total_qty' :0, 'count': 0}

        product_data[product]['total_qty'] += sale['quantity']
        product_data[product]['count'] += 1

    average = {product: data['total_qty']/ data['count']
                  for product, data in product_data.items()}
    return average

avg_report = product_averages(sales_data)
print("\n VAZIFA 5: O'rtacha sotuvlar:")
for product, avg in avg_report.items():
    print(f"  {product}: {avg:.1f} dona/kun")
    
```
