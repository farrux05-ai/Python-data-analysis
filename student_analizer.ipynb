```python
# KUN 2: STUDENT GRADE ANALYZER
# Maqsad: zip, enumerate, map, filter MASTERY

# Data - 3 subjects 10 students
students = ['Ali', 'Vali', 'Guli', 'Dilshod', 'Madina', 
            'Sardor', 'Laylo', 'Jasur', 'Nilufar', 'Bekzod']

math_scores = [85, 92, 78, 95, 88, 73, 90, 82, 91, 87]
physics_scores = [80, 88, 82, 90, 85, 70, 92, 79, 89, 84]
chemistry_scores = [75, 90, 80, 92, 87, 68, 88, 81, 90, 86]
```

```python
# ============================================
# VAZIFA 1: Stics students and grades
# You will learn: zip() - sticking parallel list
# ============================================
def zipping_list(name, math, physics, chemistry):
    """ This function zip student list with grade"""
    records = []
    for n,m,p,ch in zip(name, math, physics, chemistry):
        records.append({
            "Students" : n,
            "Math" : m,
            "Physics" : p,
            "Chemistry" : ch
        })
    return records
student_records = zipping_list(students, math_scores, physics_scores, chemistry_scores)
print(f"Vazifa bajarildi!")
print(f"Jami talabalar: {len(student_records)}")
print(f"Birinchi talaba: {student_records[0]}")
```

```python
# ============================================
# Challange 2: Calculate average score of per student
# O'rganasiz: map() + lambda
# ============================================
def calculator_per_stundent(records):
    """ This function calculate per students's average score"""
    
    for record in records:
        scores = [record['Math'], record['Physics'], record["Chemistry"]]
        record['average'] = sum(scores)/ len(scores)

    return records
student_records = calculator_per_stundent(student_records)
print("\n Complated 2: Calculated average grade")
for student in student_records[:3]:
    print(f" {student['Students']} : {student['average']:.1f}")

    
```

```python
# ============================================
# VAZIFA 3: A'lo talabalarni ajrating (average >= 85)
# O'rganasiz: filter() + lambda
# ============================================
def excelent_student(records):
    """ This function give student who's average score equal/higher 85"""
    score_85 = list(filter(lambda x: x['average'] >= 85, records))
    return score_85
    
excelent = excelent_student(student_records)
print(f'Complated 3.  High score student {len(excelent)}') 
for student in excelent:
    print(f"Name: {student["Students"]} : {student["average"]:.1f} score")
```

```python
# ============================================
# VAZIFA 4: Reytingni yarating (enumerate bilan)
# O'rganasiz: enumerate() - index bilan ishlash
# ============================================
def create_ranking(records):
    """ This functions create rankings for students"""
    sorted_st = sorted(records, key=lambda x: x['average'], reverse=True)

    ranking = []
    for rank, sorting in enumerate(sorted_st, start=1):
        ranking.append({
            "rank" : rank,
            "name" : sorting['Students'],
            "average" : sorting['average']
        })
    return ranking
ranking = create_ranking(student_records)
print("\n VAZIFA 4: Reyting jadvali:")
print("Rank  | Name      | Average")
print("-" * 30)
for student in ranking:
    print(f"{student['rank']} | {student['name']} |{student['average']:.1f}") 
```

```python
# ============================================
# VAZIFA 7: Har bir talabaning eng yaxshi fanini aniqlang
# O'rganasiz: max on dict values
# ============================================

def best_subject_per_student(records):
    """Har bir talaba uchun eng yaxshi fanini topadi"""
    # SIZNING KODINGIZ
    
    results = []
    for student in records:
        subjects = {
            'Math': student['math'],
            'Physics': student['physics'],
            'Chemistry': student['chemistry']
        }
        best = max(subjects, key=subjects.get)
        results.append({
            'name': student['name'],
            'best_subject': best,
            'score': subjects[best]
        })
    
    return results

best_subjects = best_subject_per_student(student_records)
print("\n VAZIFA 7: Talabalarning eng yaxshi fanlari:")
for student in best_subjects[:5]:
    print(f"  {student['name']:10} → {student['best_subject']:10} ({student['score']})")

```

```python
# ============================================
# Kuchsiz talabalarni topish va yordam berish
# ============================================

def students_need_help(records, threshold=75):
    """Qaysi talabalar qaysi fanlarda yordamga muhtoj"""
    # SIZNING KODINGIZ
    
    needs_help = []
    
    for student in records:
        weak_subjects = []
        if student['math'] < threshold:
            weak_subjects.append(('Math', student['math']))
        if student['physics'] < threshold:
            weak_subjects.append(('Physics', student['physics']))
        if student['chemistry'] < threshold:
            weak_subjects.append(('Chemistry', student['chemistry']))
        
        if weak_subjects:
            needs_help.append({
                'name': student['name'],
                'weak_subjects': weak_subjects
            })
    
    return needs_help

help_needed = students_need_help(student_records)
print(f"\n BONUS: Yordamga muhtoj: {len(help_needed)} ta talaba")
for student in help_needed:
    subjects = ', '.join([f"{s[0]} ({s[1]})" for s in student['weak_subjects']])
    print(f"  {student['name']}: {subjects}")

```
